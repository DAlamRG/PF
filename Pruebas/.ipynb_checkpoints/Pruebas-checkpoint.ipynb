### Energy distribution as function of temperature

```julia
"""
    determineλ(lngE,T)
Given a dictionary `lngE` whose keys(values) are energies(log of energy densities), and a temperature `T`; returns the largest value
λ of the differences between energy density `ln(g(E))` and `βE` (i.e. `max{log(g(Eᵢ))-βE}`).
"""
function determineλ(lngE::Dict{Float64,Float64},T::Float64)
    β = 1/T
    difs = Float64[]
    for el in lngE
        e,lnge = el 
        dif = lnge-(β*e)
        push!(difs,dif)
    end
    λ = maximum(difs)
    return λ
end


"""
    sExp(E,T,lnge,λ)
Given a temperature `T`, an energy `E`, the logarithm of the enegy density `lnge` and a number `λ`; returns e^{ln(g(e))-βE-λ}.
"""
function sExp(E,T,lnge,λ)
    β = 1/T
    res = exp(lnge-β*E-λ)
    return res
end



function plt_energyDist(protein_name::String,model_name::String,geometry::String,temperatures)
    pathname = "/Volumes/Sims/Resultados_sims/"*protein_name*"/"*model_name
   
    pathWL =  pathname*"/WL_"*protein_name*"_"*model_name*"_"*geometry
    
    lngE_aux  = readdlm(pathWL*"/lngE.csv",',')
    lngE = Dict{Float64,Float64}(lngE_aux[i,1] => lngE_aux[i,2] for i in 1:length(lngE_aux[:,1]))
    energies = sort(Float64[e for e in keys(lngE)]) 
    
    nTs = length(temperatures)
    nEs = length(energies)
    ps = zeros(Float64,(nEs,nTs)) # Array contains the value of the partition function for all of the wanted temps.
    
    for i in 1:nTs
        T = temperatures[i]
        β = 1/T
        𝚭 = 0
        λ = determineλ(lngE,T)
        for e in energies # Compute the value of the partition function at the given temperature.
            lnge = lngE[e]
            z = sExp(e,T,lnge,λ)
            𝚭 = 𝚭 + z
        end
        for j in 1:nEs
            ej = energies[j]
            expj = sExp(ej,T,lngE[ej],λ)
            ps[j,i] = expj/𝚭
        end
    end
    
    temp1 = temperatures[1]
    plt = plot(energies, ps[:,1], lw = 1.8, alpha = 0.8, xlabel=L"E", 
        ylabel=L"P(H(\mathbf{\sigma})=E)",label=L"T= "*" $temp1",
        legend=:topright, xguidefontsize = 15, yguidefontsize = 15, legendfontsize = 13, xtickfontsize = 14, ytickfontsize = 13, titlefontsize = 14.6)
    xlims!((energies[1]-0.1,energies[end]+0.1))
    for k in 2:nTs
        tempk = temperatures[k]
        plot!(energies, ps[:,k], lw = 1.8, alpha = 0.8,label=L"T= "*" $tempk",
        legend=:topright, xguidefontsize = 15, yguidefontsize = 15, legendfontsize = 13, xtickfontsize = 14, ytickfontsize = 13, titlefontsize = 14.6)
    end
    return plt
end
```

## Minimum config plots

```julia
"""
    triangularLattice(nα,nβ)
Returns a matrix with the coordinates for `nα*nβ` points in a triangular lattice.
"""
function triangularLattice(nα,nβ)
    v1 = [1,0]
    v2 = [-1/2,sqrt(3)/2] # Basis vectors for the triangular Bravais lattice.
    
    bM = zeros(Float64,(Int(nα*nβ),2)) # This matrix will contain the linear combinations.
    
    for β in 1:nβ, α in 1:nα
        v = α*v1 + β*v2
        k = α + (β-1)*nα
        bM[k,:] = v
    end
    return bM
end








"""
    chBasisTriangular(m)
Given a matrix `m` whose rows are the coordinates in a regular square lattice; returns the same elements but in the triangular lattice.
"""
function chBasisTriangular(m)
    A = [[-1/2 1];[sqrt(3)/2 0]] # Change of basis matrix.
    mp = zeros(Float64,size(m)) # This array contains the vectors expressed in the new basis.
    for k in 1:size(m)[1]
        el = m[k,:]
        nv = A*el
        mp[k,:] = nv
    end
    return mp
end




"""
   fccLattice(n)  
Given the size of a 3D array `n`; returns a matrix whose rows are the positions on the fcc lattice..
"""
function fccLattice(n)
    a1 = [1,1,0]
    a2 = [0,1,1] 
    a3 = [1,0,1] # Basis vectors for the Bravais lattice (fcc)
    
    fccM = zeros(Float64,(Int(n^3),3)) # This matrix will contain the linear combinations. This is the fcc lattice. 
    # Each entry fccM[i,j,k] represents the site \vec{r}=(i,k,k).
    
    for k in 1:n, j in 1:n, i in 1:n
        v = i*a1 + j*a2 + k*a3
        l = j + (k-1)*n
        s = i + (l-1)*n
        fccM[s,:] = v
    end
    return fccM
end







"""
   fccPositions(N,edo,HPlist)  
Given a 3D array size `N`, a configuration `edo`, and a sequence of aminoacids `HPlist`; returns the configuration in the fcc lattice.
"""
function fccPositions(N,edo,HPlist)
    a1 = [1,1,0]
    a2 = [0,1,1] 
    a3 = [1,0,1] # Basis vectors for the Bravais lattice (fcc)
    
    edofcc = ones(Float64,(length(HPlist),3)) # This matrix will contain the positions in the fcc lattice.

    for w in 1:length(HPlist) # Iterate over the monomers.
        inds = edo[w,:]
        i,j,k = inds # Position in the previous array.
        i = mod1(i,N)
        j = mod1(j,N)
        k = mod1(k,N) # Make the indices periodic.
        v = i*a1 + j*a2 + k*a3
        edofcc[w,:] = v
    end
    return edofcc
end




"""
    color_amin(HPlist)

Given a list of aminoacids `HPlist`, returns a list ofthe corresponding colours. 
"""
function color_amin(HPlist)
    colours = RGB{Float64}[]

    for i in 1:length(HPlist)
        el = HPlist[i]
        if el == 1
            push!(colours,RGB{Float64}(0.5,0.4,0.8))
        elseif el == 2
            push!(colours,RGB{Float64}(0.85,0.2,0.3))
        elseif el == 3
            push!(colours,RGB{Float64}(0.5,0.4,0.8))
        elseif el == 4
            push!(colours,RGB{Float64}(0.6,0.9,0.65))
        elseif el == 5
            push!(colours,RGB{Float64}(0.5,0.6,0.75))
        elseif el == 6
            push!(colours,RGB{Float64}(0.5,0.4,0.8))
        end
    end
    return colours
end




function vis_min(protein_name::String,geometry::String,camarg)
    pathname = "/Volumes/Sims/Resultados_sims/"*protein_name*"/HPNX/WL_"*protein_name*"_HPNX_"*geometry
    
    edo = readdlm(pathname*"/min_edo.csv",',')
    HPlist = vec(readdlm(pathname*"/HPlist.csv"))
    colours = color_amin(HPlist)
    
    N = Int(vec(readdlm(pathname*"/latticesize.csv",','))[1])
    

    if geometry == "square"
        plt = plot(edo[:,2],edo[:,1],lw=2,markershape=:circle,markercolor=colours ,markersize=7,color="purple",label="",xlabel="",
        ylabel="",xguidefontsize=15,yguidefontsize=15,legendfontsize=12,xtickfontsize=14.0,ytickfontsize=14.0,titlefontsize=15)
        xlims!((minimum(edo[:,2])-1,maximum(edo[:,2])+1))
        ylims!((minimum(edo[:,1])-1,maximum(edo[:,1])+1))

    
    elseif geometry == "triangular"
        tl = triangularLattice(N,N) # Generate the background lattice
        edo = chBasisTriangular(edo) # Change the basis for the current configuration `edo`.

        plt = scatter(tl[:,1],tl[:,2],color="gray",alpha=0.01,label="",markersize=7,grid=:false,xlabel="",
        ylabel ="",xguidefontsize=15,yguidefontsize=15,legendfontsize=12,xtickfontsize=14.0,ytickfontsize=14.0,titlefontsize=15)
        plot!(edo[:,1],edo[:,2],lw=2,markershape=:circle,markercolor=colours ,markersize=8,color="purple",label="",linealpha=0.8)
        xlims!((minimum(edo[:,1])-0.2,maximum(edo[:,1])+0.2))
        ylims!((minimum(edo[:,2])-0.2,maximum(edo[:,2])+0.2))
    
    
    elseif geometry == "fcc"
        #xs = range(-0.5,0.5, length=100)
        #ys = range(-0.5,0.5, length=100)
        
        edo = fccPositions(N,edo,HPlist)
        
        plt = plot(edo[:,1],edo[:,2],edo[:,3],lw=2,markershape=:circle,markercolor=colours,camera = camarg,
        markersize=8,color="purple",label="",grid=:true,xlabel="",ylabel="",zlabel="", xguidefontsize=15,yguidefontsize=15,
        legendfontsize=12,xtickfontsize=13.2,ytickfontsize=13.2,ztickfontsize=13.2,titlefontsize=14,aspect_ratio =:equal)
        #for i in 1:length(HPlist)
         #   surface!(edo[i,1].+xs,edo[i,1].+ys,(x, y)->sqrt(1-(x^2+y^2)) +edo[i,3])
        #end
        xlims!((minimum(edo[:,1])-1,maximum(edo[:,1])+1))
        ylims!((minimum(edo[:,2])-1,maximum(edo[:,2])+1))
        zlims!((minimum(edo[:,3])-1,maximum(edo[:,3])+1))
    end

    return plt

end
```

## Simple exact HP model graphs

```julia
using Plots
using LaTeXStrings
```

```julia
function u(T)
    β = 1/T
    eb = exp(β)
    return - (eb)/(4(4+eb))
end

function c(T)
    β = 1/T
    eb = exp(β)
    a = 4*(β^2)*eb
    b = (4+eb)^2
    return a/(4*b)
end

function S(T)
    β = 1/T
    eb = exp(β)
    a = - (β*eb)/(4+eb)
    b = log(4+eb)
    return a+b
end

function F(T)
    β = 1/T
    eb = exp(β)
    a = (-T)*log(4+eb)
    return a
end
```

```julia
pgfplotsx()
function pltu(ti,tf,nTs)
    temps = range(ti,stop=tf, length=nTs)
    us = Float64[u(T) for T in temps]
    plt=plot(temps,us,lw=2.8,color=RGB{Float64}(0.8,0.2,0.3),xlabel=L"$T$",ylabel=L"\langle E \rangle/N",label="",title="",
    alpha=0.8,legend=:topleft,xguidefontsize=15,yguidefontsize=15,legendfontsize=12,xtickfontsize=13,ytickfontsize=13)
end

function pltc(ti,tf,nTs)
    temps = range(ti,stop=tf, length=nTs)
    cs = Float64[c(T) for T in temps]
    plt=plot(temps,cs,lw=2.8,color=RGB{Float64}(0.6,0.4,0.85),xlabel=L"$T$",ylabel=L"C_{V}/N",label="",title="",
    alpha=0.8,legend=:topleft,xguidefontsize=15,yguidefontsize=15,legendfontsize=12,xtickfontsize=13,ytickfontsize=13)
end

function pltS(ti,tf,nTs)
    temps = range(ti,stop=tf, length=nTs)
    Ss = Float64[S(T) for T in temps]
    plt=plot(temps,Ss,lw=2.8,color="purple",xlabel=L"$T$",ylabel=L"$S(T)$",label="",title="",
    alpha=0.65,legend=:topleft,xguidefontsize=15,yguidefontsize=15,legendfontsize=12,xtickfontsize=13,ytickfontsize=13)
end

function pltF(ti,tf,nTs)
    temps = range(ti,stop=tf, length=nTs)
    Fs = Float64[F(T) for T in temps]
    plt=plot(temps,Fs,lw=2.8,color="green",xlabel=L"$T$",ylabel=L"$F(T)$",label="",title="",
    alpha=0.65,legend=:topleft,xguidefontsize=15,yguidefontsize=15,legendfontsize=12,xtickfontsize=13,ytickfontsize=13)
end
```

```julia
using LinearAlgebra
function plt_cv_prog(model_name::String,geometry::String)
    pathname1 = "/Volumes/Sims/Resultados_sims/"*"chignolin"*"/"*model_name
    pathname2 = "/Volumes/Sims/Resultados_sims/"*"trp_cage"*"/"*model_name
    pathname3 = "/Volumes/Sims/Resultados_sims/"*"villin"*"/"*model_name
   
    pathWL1 =  pathname1*"/WL_"*"chignolin"*"_"*model_name*"_"*geometry
    pathWL2 =  pathname2*"/WL_"*"trp_cage"*"_"*model_name*"_"*geometry
    pathWL3 =  pathname3*"/WL_"*"villin"*"_"*model_name*"_"*geometry
    #pathMET = pathname*"/MET_"*protein_name*"_"*model_name*"_"*geometry

    csWL1 = vec(readdlm(pathWL1*"/cs.csv",','))
    csWL2 = vec(readdlm(pathWL2*"/cs.csv",','))
    csWL3 = vec(readdlm(pathWL3*"/cs.csv",','))
    
    x1 = 10
    x2 = 20
    x3 = 35
    xs = [x1,x2,x3]
    
    y1 = maximum(csWL1)*x1
    y2 = maximum(csWL2)*x2
    y3 = maximum(csWL3)*x3
    ys = [y1,y2,y3]
    
    v = xs\ys
    display(v)
    m = round(v,digits=3)
    
    xr = range(0,stop=x3+2,length=100)
        
    plt = scatter(xs,ys,ms=5.8,color=RGB{Float64}(0.5,0.4,0.8),alpha=0.95,xlabel=L"N",label=L"\textrm{WL}",legend=:bottomright,
        ylabel=L"C_{V}(T_c)",xguidefontsize=15,yguidefontsize=15,legendfontsize=13,xtickfontsize=14.0,ytickfontsize=14.0,titlefontsize=14.6)
        plot!(xr,Float64[v*x for x in xr],label=L"m="*" $m",lw=2.2,color=RGB{Float64}(0.8,0.2,0.3),alpha=0.78)
        
    return plt
    
end
```

```julia
function plt_cv_prog_3()
    models = ["HP1","HPNX","Full1"]
    geoms = ["square","triangular","fcc"]
    
    xs = Float64[]
    ys = Float64[]
    
    for i in 1:3
        for j in 1:3
            model_name = models[i]
            geometry = geoms[j]
            pathname1 = "/Volumes/Sims/Resultados_sims/"*"chignolin"*"/"*model_name
            pathname2 = "/Volumes/Sims/Resultados_sims/"*"trp_cage"*"/"*model_name
            pathname3 = "/Volumes/Sims/Resultados_sims/"*"villin"*"/"*model_name
   
            pathMET1 =  pathname1*"/MET_"*"chignolin"*"_"*model_name*"_"*geometry
            pathMET2 =  pathname2*"/MET_"*"trp_cage"*"_"*model_name*"_"*geometry
            pathMET3 =  pathname3*"/MET_"*"villin"*"_"*model_name*"_"*geometry

            csMET1 = vec(readdlm(pathMET1*"/thermo_quantities/cp.csv",',')) 
            csMET2 = vec(readdlm(pathMET2*"/thermo_quantities/cp.csv",','))
            csMET3 = vec(readdlm(pathMET3*"/thermo_quantities/cp.csv",','))


            x1 = 10
            x2 = 20
            x3 = 35
            xsm = [x1,x2,x3]

            y1 = maximum(csMET1)*x1
            y2 = maximum(csMET2)*x2
            y3 = maximum(csMET3)*x3
            ysm = [y1,y2,y3]
            
            append!(xs,xsm)
            append!(ys,ysm)
            
        end
    end
    
    v = xs\ys
    m = round(v,digits=3)
    xr = range(0,stop=maximum(xs)+2,length=100)
    
    plt = scatter(xs,ys,ms=5.8,color=RGB{Float64}(0.5,0.4,0.8),alpha=0.95,xlabel=L"N",label=L"\textrm{MET}",legend=:bottomright,
                ylabel=L"C_{V}(T_c)",xguidefontsize=15,yguidefontsize=15,legendfontsize=13,xtickfontsize=14.0,ytickfontsize=14.0,titlefontsize=14.6)
                plot!(xr,Float64[v*x for x in xr],label=L"m="*" $m",lw=2.2,color=RGB{Float64}(0.8,0.2,0.3),alpha=0.78)
            
    return plt
    
end
```

```julia
pltdistya = plt_cv_prog_3()
ylims!((-1,30))
```
